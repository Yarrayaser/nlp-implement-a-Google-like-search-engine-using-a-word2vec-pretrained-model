```python
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def processing(doc1):
    # make all words in document lower
    processed_doc1 = doc1.lower()
    # Cleaing the document
    processed_doc1 = re.sub('[^a-zA-Z]', ' ', processed_doc1 )
    processed_article = re.sub(r'\s+', ' ', processed_doc1)
    # tokenization for document
    all_sentences = nltk.sent_tokenize(processed_article)
    all_words = [nltk.word_tokenize(sent) for sent in all_sentences]
    # Removing Stop Words
    from nltk.corpus import stopwords
    for i in range(len(all_words)):
        all_words[i] = [w for w in all_words[i] if w not in stopwords.words('english')]
     
    return all_words 
```

```python
#function make word2vec for each document in domains
def word2vec_domain(domain):
    from gensim.models import Word2Vec
    all_word_doc=[]
    for i in domain:
        #make word2vec model for each document in domains
        model = Word2Vec(i, min_count=10)
        #get words of document from word2vec model 
        words_doc = list(model.wv.key_to_index)
        all_word_doc.append(words_doc)
    return(all_word_doc)
```

```python
import numpy as np
#function to make average method for all words in document
def average_method(doc):
    embeded_vec_doc=np.average(doc, axis=0)
    return embeded_vec_doc
```

```python
from gensim import models
#get pretrained model
w = models.KeyedVectors.load_word2vec_format(
    '../input/nlpword2vecembeddingspretrained/GoogleNews-vectors-negative300.bin', binary=True)
```
